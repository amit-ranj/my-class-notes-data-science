# 1. VARIABLE IDENTIFICATION

- We need to identify dependent and independent variables.
- **dependent variables (dv) aka target variable aka predictive variable = y**
- **independent variables (idv) aka non target variable aka non predictive variable aka features = x**
- there should be one dependent variable  and multiple independent variable
- Machine learning is devided into 3 parts
  - Regression -> if dv is continuous
  - classification -> if dv is binary or category
  - clustering -> if no dv is present

**1. variable indentification**

While building any ML project we always use only relevant attributes.
Using ir-relevant attributes results in Multi-Collinearity or Overfitting problem.

Multicollinearity refers to a situation where two or more independent variables in a dataset are highly correlated with each other. This can create problems in regression models, making it difficult to determine the individual effect of each variable.

# 5. missing value treatment
- in dataset, we have:-

  - number (numerical data) -> if this is missing then,

                              we fill the gap by the mean / median / mode of data.
    
                                mean strategy

                                median strategy

                                mode strategy      

  - text (categorical data) -> if this is missing then,
    
                                  we fill the gap by mode of data.

                                          if mode is not present then, fill data based on the nearest value in other attributes related to it. (nearest neighbour algorithms)
