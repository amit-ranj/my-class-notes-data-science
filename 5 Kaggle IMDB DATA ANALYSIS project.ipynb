**Series**

In pandas, a Series is a one-dimensional labeled array capable of holding any data type (integers, floats, strings, Python objects, etc.). It is similar to a column in an Excel spreadsheet or a single column in a pandas DataFrame.

Each column is a panda series.

### **pandas `DataFrame` – The Core Data Structure**  
A **pandas DataFrame** is a **two-dimensional table** similar to an **Excel spreadsheet** or a **SQL table**. It consists of **rows and columns**, where:  
- **Rows** have an index (default is numeric but can be customized).  
- **Columns** have labels (column names).  
- Can hold **different data types** (integers, floats, strings, etc.).  

---

## **Creating a DataFrame**
You can create a pandas **DataFrame** from various sources like lists, dictionaries, NumPy arrays, CSV files, SQL databases, etc.

### **1. Creating a DataFrame from a Dictionary**
```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Salary': [50000, 60000, 70000]
}

df = pd.DataFrame(data)
print(df)
```
**Output:**
```
      Name  Age  Salary
0   Alice   25  50000
1     Bob   30  60000
2  Charlie   35  70000
```
- Each **column** is a **pandas Series**.
- The **index** (0, 1, 2) is automatically generated.

---

### **2. Creating a DataFrame with Custom Index**
```python
df = pd.DataFrame(data, index=['A', 'B', 'C'])
print(df)
```
**Output:**
```
      Name  Age  Salary
A   Alice   25  50000
B     Bob   30  60000
C  Charlie   35  70000
```
- The custom **index labels** (`A`, `B`, `C`) replace the default numeric index.

---

### **3. Creating a DataFrame from a List of Lists**
```python
data = [
    ['Alice', 25, 50000],
    ['Bob', 30, 60000],
    ['Charlie', 35, 70000]
]

df = pd.DataFrame(data, columns=['Name', 'Age', 'Salary'])
print(df)
```
- `columns` parameter assigns column names.

### **`movies.isnull().any().any()` in pandas**  

This command is used to **check if there are any missing (`NaN`) values** in a DataFrame.

---

### **1. Breaking Down the Command**
| Command | Explanation |
|---------|------------|
| `movies.isnull()` | Returns a **DataFrame** of the same shape as `movies`, with `True` for missing values (`NaN`) and `False` otherwise. |
| `movies.isnull().any()` | Returns a **Series**, with `True` for columns that contain at least one missing value. |
| `movies.isnull().any().any()` | Returns a **single Boolean value (`True` or `False`)** indicating if **any column has missing values** in the entire DataFrame. |

---

### **2. Example: Checking for Missing Values**
```python
import pandas as pd
import numpy as np

# Sample dataset with missing values
data = {
    'movie_id': [101, 102, 103, 104, 105],
    'title': ['Inception', 'Titanic', 'Avatar', np.nan, 'Interstellar'],
    'rating': [8.8, 7.8, np.nan, 9.0, 8.6]
}

movies = pd.DataFrame(data)

# Check if any missing values exist
print(movies.isnull().any().any())
```

#### **Output:**
```
True
```
This means that **at least one missing value (`NaN`) exists in the DataFrame**.

---

### **3. Alternative Methods to Detect Missing Values**
#### **Method 1: Check for Missing Values in Each Column**
```python
print(movies.isnull().sum())
```
**Output:**
```
movie_id    0
title       1
rating      1
dtype: int64
```
This shows that:
- **1 missing value in `title`**
- **1 missing value in `rating`**
- **0 missing values in `movie_id`**

---

#### **Method 2: Get Total Number of Missing Values**
```python
print(movies.isnull().sum().sum())
```
**Output:**
```
2
```
This tells us there are **2 missing values** in total.

---

### **4. How to Handle Missing Values?**
#### **A. Remove Missing Values**
```python
movies_cleaned = movies.dropna()
```
✅ Removes all rows with `NaN`.

#### **B. Fill Missing Values (Imputation)**
```python
movies_filled = movies.fillna({'title': 'Unknown', 'rating': movies['rating'].mean()})
```
✅ Replaces missing `title` with `"Unknown"` and missing `rating` with the **average rating**.

---

### **5. Summary**
| Command | Output |
|---------|--------|
| `movies.isnull().any().any()` | `True` if **any** missing value exists, `False` otherwise |
| `movies.isnull().sum()` | Count of missing values per column |
| `movies.isnull().sum().sum()` | Total number of missing values |
| `movies.dropna()` | Removes rows with missing values |
| `movies.fillna(value)` | Replaces missing values |

### **`tags.dropna()` in pandas**  
The `dropna()` function is used to **remove rows or columns** that contain **missing values (`NaN`)**.

---

### **1. How `tags.dropna()` Works**
By default, `dropna()` **removes rows** that contain at least **one `NaN` value**.

```python
tags.dropna()
```
✅ **Removes all rows** where any column has a missing value.

---

### **2. Example: Removing Missing Values from `tags` DataFrame**
```python
import pandas as pd
import numpy as np

# Sample dataset with missing values
data = {
    'tag_id': [1, 2, 3, 4, 5],
    'movie_id': [101, 102, np.nan, 104, 105],
    'tag': ['Action', 'Drama', np.nan, 'Sci-Fi', 'Comedy']
}

tags = pd.DataFrame(data)

print("Original DataFrame:")
print(tags)

# Drop rows with NaN values
tags_cleaned = tags.dropna()

print("\nDataFrame after dropna():")
print(tags_cleaned)
```

#### **Output:**
**Original DataFrame:**
```
   tag_id  movie_id     tag
0       1     101.0  Action
1       2     102.0   Drama
2       3       NaN     NaN
3       4     104.0  Sci-Fi
4       5     105.0  Comedy
```
**After `dropna()`:**
```
   tag_id  movie_id     tag
0       1     101.0  Action
1       2     102.0   Drama
3       4     104.0  Sci-Fi
4       5     105.0  Comedy
```
🚀 **Row 2 (index 2) was removed because it contained `NaN` values.**

---

### **3. Different Ways to Use `dropna()`**
#### **A. Remove Rows with `NaN` (Default Behavior)**
```python
tags.dropna()
```
✅ Removes **any row** with at least **one `NaN`**.

#### **B. Remove Only Columns with `NaN`**
```python
tags.dropna(axis=1)
```
✅ Removes **entire columns** that have any missing values.

#### **C. Remove Rows Only If All Values Are `NaN`**
```python
tags.dropna(how='all')
```
✅ Removes **only those rows** where **all columns are `NaN`**.

#### **D. Remove Rows If a Specific Column Has `NaN`**
```python
tags.dropna(subset=['movie_id'])
```
✅ Removes rows where **`movie_id`** is `NaN`, but keeps other rows.

#### **E. Keep Rows with At Least `n` Non-NaN Values**
```python
tags.dropna(thresh=2)
```
✅ Keeps rows that have at least **2 non-null values**.

---

### **4. Alternative: Fill Missing Values Instead**
If you don't want to remove data, you can **fill `NaN` values**:
```python
tags.fillna({'movie_id': 0, 'tag': 'Unknown'})
```
✅ Replaces missing `movie_id` with `0` and missing `tag` with `"Unknown"`.

---

### **5. Summary**
| Command | Effect |
|---------|--------|
| `tags.dropna()` | Removes rows with **any** `NaN` |
| `tags.dropna(axis=1)` | Removes columns with `NaN` |
| `tags.dropna(how='all')` | Removes rows **only if all values are `NaN`** |
| `tags.dropna(subset=['movie_id'])` | Removes rows where `movie_id` is `NaN` |
| `tags.dropna(thresh=2)` | Keeps rows with at least `2` non-null values |
| `tags.fillna(value)` | **Replaces** missing values instead of removing |

# Vectorized string operations

### **Vectorized String Operations in Pandas**
Pandas provides **vectorized string operations** using the `.str` accessor, which allows efficient manipulation of text data stored in a **Series** or **DataFrame column**.

---

## **🔹 Key Vectorized String Functions**
Instead of applying loops, you can use `.str` methods to perform operations efficiently.

### **✅ 1. Converting to Lowercase & Uppercase**
```python
df['column'].str.lower()  # Convert to lowercase
df['column'].str.upper()  # Convert to uppercase
df['column'].str.title()  # Convert to title case (first letter capitalized)
```

🔹 **Example**
```python
import pandas as pd

data = {'movie': ['The Matrix', 'Inception', 'AVATAR']}
df = pd.DataFrame(data)

df['movie_lower'] = df['movie'].str.lower() # add column movie_lower and store in lowercase
df['movie_upper'] = df['movie'].str.upper()
print(df)
```

#### **📌 Output**
```
       movie     movie_lower    movie_upper
0  The Matrix  the matrix      THE MATRIX
1  Inception   inception      INCEPTION
2  AVATAR      avatar         AVATAR
```

---

### **✅ 2. String Length**
```python
df['column'].str.len()
```
🔹 **Example**
```python
df['title_length'] = df['movie'].str.len()
```

---

### **✅ 3. Checking for a Substring**
```python
df['column'].str.contains('word', case=False, na=False)
```
🔹 **Example**
```python
df['has_avatar'] = df['movie'].str.contains('avatar', case=False)
```
This checks if **'avatar'** is present in the **movie** column.

---

### **✅ 4. Replacing Substrings**
```python
df['column'].str.replace('old', 'new', regex=True)
```
🔹 **Example**
```python
df['movie_fixed'] = df['movie'].str.replace('AVATAR', 'Avatar')
```

---

### **✅ 5. Splitting Strings**
```python
df['column'].str.split('delimiter')
```
🔹 **Example**
```python
df['movie_words'] = df['movie'].str.split(' ')
```

---

### **✅ 6. Extracting Substrings Using Regex**
```python
df['column'].str.extract(r'pattern')
```
🔹 **Example: Extracting year from `"The Matrix (1999)"`**
```python
df['year'] = df['movie'].str.extract(r'\((\d{4})\)')
```

---

### **✅ 7. Removing Whitespace**
```python
df['column'].str.strip()  # Removes leading & trailing spaces
df['column'].str.lstrip() # Removes leading spaces
df['column'].str.rstrip() # Removes trailing spaces
```

---

### **✅ 8. Concatenating Strings**
```python
df['new_column'] = df['column1'].str.cat(df['column2'], sep=' ')
```

---

### **Summary**
| Operation | Function |
|-----------|----------|
| Convert Case | `.str.lower()`, `.str.upper()`, `.str.title()` |
| String Length | `.str.len()` |
| Contains Substring | `.str.contains('text', case=False)` |
| Replace Substring | `.str.replace('old', 'new')` |
| Split Strings | `.str.split(' ')` |
| Extract Substring (Regex) | `.str.extract(r'pattern')` |
| Remove Whitespace | `.str.strip()` |
| Concatenation | `.str.cat()` |
