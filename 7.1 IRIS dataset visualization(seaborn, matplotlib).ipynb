### BoxPLot or Whisker Plot
A boxplot is used to visualize the distribution of numerical data and identify outliers. It displays the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum values of a dataset.

```python
iris.head()
```

```python
fig = plt.gcf()
fig.set_size_inches(10,7)
fig = sns.boxplot(x = 'Species', y = 'PetalLengthCm', data = iris, order = ['Iris-virginica','Iris-versicolor','Iris-setosa'],linewidth=2.5, orient = 'v', dodge = False, palette={'Iris-virginica': 'red', 'Iris-versicolor': 'green', 'Iris-setosa': 'blue'})
plt.show()
```

**plt.gcf()** â†’ Gets the current figure (gcf = "Get Current Figure") from Matplotlib.

**ðŸ“Œ What is gcf() in Matplotlib?**

plt.gcf() stands for "Get Current Figure" in Matplotlib. It retrieves the current figure object so you can modify it (e.g., resizing, adding elements).

**ðŸ“Œ What is plt.gca() in Matplotlib?**

plt.gca() stands for "Get Current Axes" in Matplotlib. It retrieves the current Axes (subplot) object, allowing you to modify the title, labels, grid, limits, etc. of the active plot.

**fig.set_size_inches(10,7)** â†’ Resizes the figure to 10 inches wide and 7 inches tall, making the plot larger.

**sns.boxplot(x = 'Species', y = 'PetalLengthCm', data = iris,...)** â†’ Creates a boxplot.
- x='Species' â†’ The categorical variable on the x-axis.
- y='PetalLengthCm' â†’ The numerical variable on the y-axis (Petal Length).
- data=iris â†’ Uses the iris dataset.

**order=['Iris-virginica', 'Iris-versicolor', 'Iris-setosa']**
- Specifies the order of species on the x-axis (default order is alphabetical, but here it's customized).

**linewidth=2.5**

Increases the thickness of the box edges.

**orient='v'**

Specifies vertical orientation ('v' for vertical, 'h' for horizontal).

**dodge=False**

Does nothing here, as dodge is used when multiple groups are plotted side by side (e.g., when using hue=).

If dodge=True, multiple boxplots in the same category would be separated slightly.

**palette={'Iris-virginica': 'red', 'Iris-versicolor': 'green', 'Iris-setosa': 'blue'}**

Assigns specific colors to each box based on the species.

```python
iris.boxplot(by = "Species", figsize = (12,6))
plt.show()
```

**iris.boxplot(by="Species")**

Creates separate boxplots for each numerical column in the dataset, grouped by Species.

Each subplot will show the distribution of a feature (e.g., SepalLengthCm, SepalWidthCm, etc.) for different species.

**figsize=(12,6)**

Sets the figure size to 12 inches wide and 6 inches tall, making the plots larger and more readable.

### Strip plot

A strip plot is similar to a scatter plot but is used for categorical data. It shows individual data points along with a categorical variable, helping visualize the distribution and density of values.


```python
fig = plt.gcf()
fig.set_size_inches(10,7)
fig = sns.stripplot(x = 'Species', y = 'SepalLengthCm', data = iris, jitter = True, edgecolor = 'gray', size = 8, palette = 'winter', orient = 'v')
plt.show()
```



The orient parameter in Seaborn controls the orientation of the plot:

- orient='v' â†’ Vertical plot (default).


  ðŸ”¹ Here,
  - species is on the X-axis (categorical).
  - sepal_length is on the Y-axis (numeric).
  - Dots are plotted vertically.

- orient='h' â†’ Horizontal plot.


  ðŸ”¹ Here,
  - species is now on the Y-axis.
  - sepal_length is on the X-axis.
  - Dots are plotted horizontally.



```python
fig = sns.stripplot(x = 'Species', y = 'SepalLengthCm', data = iris, jitter = True, edgecolor = 'gray', size = 8, palette = 'winter', orient = 'h')
plt.show()
```

### Combining Box and Strip Plots

```python
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.boxplot(x='Species',y='SepalLengthCm',data=iris)
fig=sns.stripplot(x='Species',y='SepalLengthCm',data=iris,jitter=True,edgecolor='gray')
plt.show()
```

**fig is Overwritten**
- fig = sns.boxplot(...) overwrites the previous fig.
- Then fig = sns.stripplot(...) overwrites the box plot.
- This does not break the code because Seaborn automatically plots everything on the same figure.



```python
ax= sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
ax= sns.stripplot(x="Species", y="PetalLengthCm", data=iris, jitter=True, edgecolor="gray")
# ax is overwritten
boxtwo = ax.artists[2] # Select the third box (index 2)
boxtwo.set_facecolor('yellow') # Change the box color to yellow
boxtwo.set_edgecolor('black') # Change the edge color to black
boxthree=ax.artists[1]
boxthree.set_facecolor('red')
boxthree.set_edgecolor('black')
boxthree=ax.artists[0]
boxthree.set_facecolor('green')
boxthree.set_edgecolor('black')

plt.show()
```

### Violin PLot
A violin plot combines a box plot and a density plot, showing:
- Median & quartiles (like a boxplot).
- Distribution shape (like a KDE plot).
- Width of the violin represents data density at different values

It is used to visualize the distribution of data and its probability distribution.This chart is a combination of a Box Plot and a Density Plot that is rotated and placed on each side, to show the distribution shape of the data. The thick black bar in the centre represents the interquartile range, the thin black line extended from it represents the 95% confidence intervals, and the white dot is the median.Box Plots are limited in their display of the data, as their visual simplicity tends to hide significant details about how values in the data are distributed

```python
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.violinplot(x='Species',y='SepalLengthCm',data=iris)
plt.show()
```

```python
plt.figure(figsize=(15,10)) 
# Creates a new figure with a size of 15x10 inches.
# Ensures that all four violin plots fit without overlapping.
plt.subplot(2,2,1)
# plt.subplot(2,2,1) â†’ Creates the first subplot (2 rows, 2 columns, 1st position).
sns.violinplot(x='Species',y='PetalLengthCm',data=iris)
# sns.violinplot() â†’ Plots Petal Length for each species.
plt.subplot(2,2,2)
# plt.subplot(2,2,2) â†’ Second subplot (2 rows, 2 columns, 2nd position) (top-right).
sns.violinplot(x='Species',y='PetalWidthCm',data=iris)
# Plots Petal Width distribution.
plt.subplot(2,2,3)
# plt.subplot(2,2,3) â†’ Third subplot (2 rows, 2 columns, 3rd position) (bottom-left).
sns.violinplot(x='Species',y='SepalLengthCm',data=iris)
# Plots Sepal Length distribution.
plt.subplot(2,2,4)
# plt.subplot(2,2,4) â†’ Fourth subplot (2 rows, 2 columns, 4th position) (bottom-right).
sns.violinplot(x='Species',y='SepalWidthCm',data=iris)
# Plots Sepal Width distribution.
plt.show()
```
